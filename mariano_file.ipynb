```python
from sklearn.decomposition import SparsePCA
```

#Since the data is sparse (maybe we should do a check before), we perform sparse PCA 


```python
# Create Sparse PCA object
sparse_pca = SparsePCA(n_components=3, alpha=0.1, random_state=6942)  # alpha is equivalent to lambda in the objective function

# Fit Sparse PCA to the data
sparse_pca.fit(df)

# Transform the data to the new reduced dimensionality
df_sparse_pca = sparse_pca.transform(df)
```

#We did it with three new dimensions and we plot the new dataset

```python
from mpl_toolkits.mplot3d import Axes3D

# Create a 3D plot
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot for the first three principal components
ax.scatter(df_sparse_pca[:, 0], df_sparse_pca[:, 1], df_sparse_pca[:, 2], alpha=0.6)

ax.set_title('Sparse PCA - 3D Scatter Plot of Transformed Data')
ax.set_xlabel('Principal Component 1')
ax.set_ylabel('Principal Component 2')
ax.set_zlabel('Principal Component 3')

plt.show()
```

#We compute explained variance for each component of the PCA

```python
# Fit Sparse PCA to the data
sparse_pca.fit(df)

# Calculate the total variance of the original data
total_variance = np.var(df)

# Compute the variance of the transformed data for each component
components_variances = np.var(sparse_pca.transform(df), axis=0)

# Calculate explained variance for each component
explained_variances = components_variances / total_variance.sum()

# Print explained variances for the first three components
for i in range(3):
    print(f"Explained Variance of Component {i+1}: {explained_variances[i]:.4f}")
```

#Since the third component does not seem very useful, we do a scatterplot just for the first two

```python
sparse_pca.fit(df.T)

# Transform the data to the new reduced dimensionality
df_sparse_pca = sparse_pca.transform(df.T)

# Plot the transformed data
plt.figure(figsize=(10, 6))
plt.scatter(df_sparse_pca[:, 0], df_sparse_pca[:, 1], alpha=0.6)
plt.title('Sparse PCA - 2D Scatter Plot of Transformed Data')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.grid(True)
plt.show()
```

#We check which of those were the most important genes for the first, second and third component

```python
# Fit Sparse PCA to the data
sparse_pca.fit(df)

# Get the loadings or coefficients
loadings = sparse_pca.components_

# Create a DataFrame to store the loadings
df_loadings = pd.DataFrame(loadings, columns=df.columns)

# Function to get top features for each principal component
def top_features(component, n=10):
    abs_component = np.abs(component)
    top_indices = np.argsort(abs_component)[-n:]
    top_features = df.columns[top_indices]
    top_loadings = abs_component[top_indices]
    return pd.DataFrame({'Feature': top_features, 'Loading': top_loadings})

# Get top features for each of the three principal components
top_features_1 = top_features(loadings[0])
top_features_2 = top_features(loadings[1])
top_features_3 = top_features(loadings[2])

print("Top features for Component 1:")
print(top_features_1)
print("\nTop features for Component 2:")
print(top_features_2)
print("\nTop features for Component 3:")
print(top_features_3)
```

#Now we would like to use the new dataset to perform Clustering

```python


# Initialize KMeans with the number of clusters (e.g., 3 clusters)
n_clusters = 4
kmeans = KMeans(n_clusters=n_clusters, random_state=4269)

# Fit KMeans to the data
kmeans.fit(df_sparse_pca_pd)

# Get cluster labels
cluster_labels = kmeans.labels_

# Plotting the clusters in 2D (using the first two principal components)
# Note: If df contains more than 2 features, you may want to use a pair of principal components or other features for plotting
plt.figure(figsize=(10, 6))
plt.scatter(df_sparse_pca_pd.iloc[:, 0], df_sparse_pca_pd.iloc[:, 1], c=cluster_labels, cmap='viridis', alpha=0.6)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', marker='x', s=100, label='Centroids')
plt.title('KMeans Clustering')
plt.xlabel(df_sparse_pca_pd.columns[0])
plt.ylabel(df_sparse_pca_pd.columns[1])
plt.legend()
plt.grid(True)
plt.show()

# Optional: Print the cluster centroids and sizes
print("Cluster Centroids:")
print(kmeans.cluster_centers_)
print("\nCluster Sizes:")
unique, counts = np.unique(cluster_labels, return_counts=True)
print(dict(zip(unique, counts)))
```
