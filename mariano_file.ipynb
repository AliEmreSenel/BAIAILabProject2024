#Since the data is sparse (maybe we should do a check before), we perform sparse PCA 


```python
# Create Sparse PCA object
sparse_pca = SparsePCA(n_components=3, alpha=0.1, random_state=6942)  # alpha is equivalent to lambda in the objective function

# Fit Sparse PCA to the data
sparse_pca.fit(df)

# Transform the data to the new reduced dimensionality
df_sparse_pca = sparse_pca.transform(df)
```

#We did it with three new dimensions and we plot the new dataset

```python
from mpl_toolkits.mplot3d import Axes3D

# Create a 3D plot
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot for the first three principal components
ax.scatter(df_sparse_pca[:, 0], df_sparse_pca[:, 1], df_sparse_pca[:, 2], alpha=0.6)

ax.set_title('Sparse PCA - 3D Scatter Plot of Transformed Data')
ax.set_xlabel('Principal Component 1')
ax.set_ylabel('Principal Component 2')
ax.set_zlabel('Principal Component 3')

plt.show()
```

#We compute explained variance for each component of the PCA

```python
# Fit Sparse PCA to the data
sparse_pca.fit(df)

# Calculate the total variance of the original data
total_variance = np.var(df)

# Compute the variance of the transformed data for each component
components_variances = np.var(sparse_pca.transform(df), axis=0)

# Calculate explained variance for each component
explained_variances = components_variances / total_variance.sum()

# Print explained variances for the first three components
for i in range(3):
    print(f"Explained Variance of Component {i+1}: {explained_variances[i]:.4f}")
```

#Since the third component does not seem very useful, we do a scatterplot just for the first two

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA()
x_new = pca.fit_transform(df)

def myplot(score,coeff,labels=None):
    xs = score[:,0]
    ys = score[:,1]
    n = coeff.shape[0]
    scalex = 1.0/(xs.max() - xs.min())
    scaley = 1.0/(ys.max() - ys.min())
    plt.scatter(xs * scalex,ys * scaley)
    for i in range(2):
        plt.arrow(0, 0, 10* coeff[i,0], 10 * coeff[i,1],color = 'r',alpha = 0.5)
        if labels is None:
            plt.text(coeff[i,0]* 1.15, coeff[i,1] * 1.15, "Var"+str(i+1), color = 'g', ha = 'center', va = 'center')
        else:
            plt.text(coeff[i,0]* 1.15, coeff[i,1] * 1.15, labels[i], color = 'g', ha = 'center', va = 'center')
plt.xlim(-1,1)
plt.ylim(-1,1)
plt.xlabel("PC{}".format(1))
plt.ylabel("PC{}".format(2))
plt.grid()

#Call the function. Use only the 2 PCs.
myplot(x_new[:,0:2],np.transpose(pca.components_[0:2, :]))
plt.show()
```

### We want to visualize Hypoxia and Normoxia in the PCA


```python
def extract_cell_name(x):
    y = x.split("_")
    return y[len(y)-2]

def get_cell_name(file_name):
    return extract_cell_name(file_name)

def convert_indexes_to_cell_names(df):
    df.index = [get_cell_name(x) for x in df.index]
    return df

def get_cell_hypo_or_norm(cell_name):
    return df_meta[df_meta["Cell name"]==cell_name]["Condition"].values[0]

def seperate_hypo_and_norm(df):
    df_hypo = df[df.index.map(lambda x: get_cell_hypo_or_norm(x)=="Hypo")]
    df_norm = df[df.index.map(lambda x: get_cell_hypo_or_norm(x)=="Norm")]
    return df_hypo, df_norm

def process_df(df):
    df = convert_indexes_to_cell_names(df)
    df_hypo, df_norm = seperate_hypo_and_norm(df)
    return df, df_hypo, df_norm
```

```python
df, df_hypo, df_norm = process_df(df.T.copy())
```

```python

def analyze_row_removal(raw_df, filt_df):
    # Identify rows that were removed
    removed_rows = raw_df.index.difference(filt_df.index)
    removed_rows_data = raw_df.loc[removed_rows]
    
    # Analyze characteristics of removed rows
    non_zero_counts = (removed_rows_data != 0).sum(axis=1)
    return removed_rows, non_zero_counts

def analyze_column_removal(raw_df, filt_df):
    # Identify columns that were removed
    removed_columns = raw_df.columns.difference(filt_df.columns)
    removed_columns_data = raw_df[removed_columns]
    
    # Analyze characteristics of removed columns
    non_zero_counts = (removed_columns_data != 0).sum(axis=0)
    return removed_columns, non_zero_counts


# Analyze row removal
removed_rows, removed_rows_non_zero_counts = analyze_row_removal(mcf7_smartS_raw, mcf7_smartS_filt)

# Analyze column removal
removed_columns, removed_columns_non_zero_counts = analyze_column_removal(mcf7_smartS_raw, mcf7_smartS_filt)

# Print results
print("Removed Rows:")
print(removed_rows)
print("\nNon-zero counts for removed rows:")
print(removed_rows_non_zero_counts)

print("\nRemoved Columns:")
print(removed_columns)
print("\nNon-zero counts for removed columns:")
print(removed_columns_non_zero_counts)

# Further analysis to deduce the rules:
# Hypothesize row removal based on threshold
row_thresholds = removed_rows_non_zero_counts.value_counts()
print("\nPossible row removal thresholds:")
print(row_thresholds)

# Hypothesize column removal based on threshold
column_thresholds = removed_columns_non_zero_counts.value_counts()
print("\nPossible column removal thresholds:")
print(column_thresholds)
```
